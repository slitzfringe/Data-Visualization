# pyplit vs matplotlib Object Oriented (OO) method
* basic and quick plotting ----> pyplot
* Advanced ----> OO method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

heart_disease = pd.read_csv("heart-disease (1).csv")
```

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head(10)
len(over_50)
```

```python
#scatter plot of age vs chol coloured based on target
over_50.plot(kind="scatter", x = "age", y='chol', c='target', figsize = (10,6));
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", ax=ax, c='target');
```

```python
# oo mrthod mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x ='age', y = 'chol', c = 'target', ax = ax);
ax.set(xlim = [45,100]);
ax.set_xlim([45,100]);

# x-axis now starts from 45-100
```

```python
over_50.target.values
```

```python
#oo method from scartch
fig, ax = plt.subplots(figsize=(10,6))

# plotting data
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target']);

#customize
ax.set(xlabel = "Age",
      ylabel = "chol",
      title = "Heart disease  and Cholestrol Levels");

#add legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(y= over_50["chol"].mean(),
          linestyle = "--");
```

```python
#lets try to make a subplots of chol, thalach(max heart rate), age
#age is  common axis

fig, (ax1, ax2) = plt.subplots(figsize=(10,10), 
                               nrows=2, 
                               ncols=1,
                              sharex=True)

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

ax1.set(title = "Age vs Chol", 
#         xlabel="age", 
        ylabel="Cholestrol")

ax1.legend(*scatter.legend_elements(),
           title = 'Target')

ax1.axhline(over_50["chol"].mean(),
            linestyle="--")



scatter = ax2.scatter(x=over_50["age"], 
                      y=over_50["thalach"],
                      c=over_50["target"]);

ax2.set(title="Age vs Thalach",
        xlabel = "Age",
        ylabel="Max heart rate")

ax2.legend(*scatter.legend_elements(), 
           title = "Target");

ax2.axhline(over_50['thalach'].mean(),
            linestyle="--");

#add a title to fig
fig.suptitle("Heart disease Analysis", fontsize = 16, fontweight = "bold") ;
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use("seaborn-whitegrid")

fig, ax = plt.subplots(figsize = (10, 6 ))
scatter = ax.scatter(x = over_50["age"],
                    y = over_50['chol'],
                    c = over_50['target'],
                    cmap = "winter") # color scheme

ax.set(title = "Heart Disease and Cholestrol Levels",
      xlabel = "Age",
      ylabel = "Cholestrol")
ax.legend(*scatter.legend_elements(), title = "Target")

ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
#customizing y and x axis limits

#lets try to make a subplots of chol, thalach(max heart rate), age
#age is  common axis

fig, (ax1, ax2) = plt.subplots(figsize=(10,10), 
                               nrows=2, 
                               ncols=1,
                              sharex=True)

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap = "winter")

ax1.set(title = "Age vs Chol", 
#         xlabel="age", 
        ylabel="Cholestrol")

ax1.set_xlim([50, 80])   #change x-axis limits

ax1.legend(*scatter.legend_elements(),
           title = 'Target')

ax1.axhline(over_50["chol"].mean(),
            linestyle="--")



scatter = ax2.scatter(x=over_50["age"], 
                      y=over_50["thalach"],
                      c=over_50["target"],
                     cmap = "summer");

ax2.set(title="Age vs Thalach",
        xlabel = "Age",
        ylabel="Max heart rate")

#ax2.set_xlim([50, 80])   #change x-axis limits
ax2.set_ylim([60,200])

ax2.legend(*scatter.legend_elements(), 
           title = "Target");

ax2.axhline(over_50['thalach'].mean(),
            linestyle="--");

#add a title to fig
fig.suptitle("Heart disease Analysis", fontsize = 16, fontweight = "bold") ;
```
